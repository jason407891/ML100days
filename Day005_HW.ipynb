```python
import requests
r=requests.get( "https://www.dcard.tw/_api/forums/pet/posts?popular=true")
re=r.text
```

```python
print(re)
```

```python
import json
data=json.loads(re)
for d in data:
 print(d['title'])
 print(d['createdAt'])
 print(d['commentCount'])
 print(d['likeCount'])
```
