```python
import requests
from bs4 import BeautifulSoup
url="https://www.ettoday.net/news/news-list.htm"
data=requests.get(url)
soup=BeautifulSoup(data.text)
s=soup.select('.part_list_2')[0].select('h3')
for i in s:
    print(i.a.text)
    print('http:/'+i.a['href'])
```
